```python
# move the previous code, once it works in a function, since we'll have to use the same code for all the csv_files
def import_month(monthlyfile):
    '''
    Read files from ex_src_path
    Add correct encoding utf-16
    Return the data
    '''
    file_path = (pathlib.Path(EX_SRC_PATH, monthlyfile))

    df = pd.read_csv(file_path, encoding="utf-16")

    return df
```

```python
# loop through the listing files
# use the glob package
# create a list of the files for the loop
monthly_files = [
    "listings3/2020/listing_1.csv",
    "listings3/2020/listing_2.csv",
    "listings3/2020/listing_3.csv",
    "listings3/2020/listing_4.csv",
    "listings3/2020/listing_5.csv",
    "listings3/2020/listing_6.csv",
    "listings3/2020/listing_7.csv",
    "listings3/2020/listing_8.csv",
    "listings3/2020/listing_9.csv",
    "listings3/2020/listing_10.csv",
    "listings3/2020/listing_11.csv",
    "listings3/2020/listing_12.csv",
    "listings3/2021/listing_1.csv",
    "listings3/2021/listing_2.csv",
    "listings3/2021/listing_3.csv",
    "listings3/2021/listing_4.csv",
    "listings3/2021/listing_5.csv",
    "listings3/2021/listing_6.csv",
    "listings3/2021/listing_7.csv",
    "listings3/2021/listing_8.csv",
    "listings3/2021/listing_9.csv"
    ]
# In order to avoid unwanted added rows
all_data_frames = []

for listing_filename in monthly_files :
  df = import_month(listing_filename)
  all_data_frames.append(df)
  # Check shapes
  print(f"{listing_filename}: {df.shape}")
```

# Compute output

```python
concat_df = pd.concat(all_data_frames, ignore_index =True)
concat_df
```

```python
# sort final_df using the price column
# check the result
price_sorted_df = concat_df.sort_values(by="price")

price_sorted_df
```

```python
# Show the first 20 rows
display(price_sorted_df.head(20))

# Show the last 20 rows
display(price_sorted_df.tail(20))

# We can't see the prices so we need to select the wanted column
```

```python
# We can't see the prices so we need to select the wanted column

display(price_sorted_df["price"].head(20))
display(price_sorted_df["price"].tail(20))

# We can see that they aren't sorted we find $0 right before $1 000 and then smaller values again
```

```python
# Type of the numbers in the price column
price_sorted_df["price"].dtype

# Objects => we can't use them as flots or integers that's why they can't be sorted properly
```

```python
# Create the new column with numbers that are floats and not objects
# Struggle to transorfm them into float. What to be aware of whan transforming here
# .replace allows to remove the problematic parts before turning into floats
# 1) the price signs as the comas prevents to turn the object into a float => remove it [\$,]
# 2) Need to tell that nothing is replacing what we remove ""
# 3) regex=True helps to get all the actual $ and , removed from all the prices
# Struggle 2 => need to double \\ in [\\$,]
price_sorted_df["true_price"] = (
    price_sorted_df["price"]
    .replace("[\\$,]", "", regex=True)
    .astype(float)
)
```

```python
# Check the type for the new column
price_sorted_df["true_price"].dtype
```

```python
#sorting by true_price
final_sorted_df = price_sorted_df.sort_values(by=["true_price"])

final_sorted_df
```

```python
# Check the first 20 rows and the last 20 rows but only the column "price" and "true_price"
# Use double brackets to see multiple columns

display(final_sorted_df[["true_price", "price"]].head(20))
display(final_sorted_df[["true_price", "price"]].tail(20))
```
