##Extract names into df_names

```python
df_names=df_people[["id","names"]].explode("names")
df_names.head()
```

##split names into given/surname

```python
def split_name(name):
  if not isinstance(name, str):
    return None, None

  name = name.strip()

  if "," in name:
    s,g=name.split(",",1)
    return g.strip(),s.strip()

  parts=name.split()
  if not parts: # Handle empty string or string with only whitespace after strip
    return None, None

  if len(parts)==1:
    return parts[0], None

  return " ".join(parts[:-1]).strip(), parts[-1].strip()

# Apply the function and convert the map object to a list of tuples
split_results = list(df_names["names"].map(split_name))

# Unpack the list of tuples into separate lists for 'given' and 'surname'
df_names["given"] = [item[0] for item in split_results]
df_names["surname"] = [item[1] for item in split_results]

df_names.head()
```

##Pick 10 random people and download their details

```python
import random, time
BASE_PERSON_URL="http://repertorium.projets.univ-poitiers.fr/rest.php/heloise/people/{}.json"
random_ids=random.sample(df_names["id"].tolist(),10)
random_ids
```

##Download details with delay 1-3 seconds

```python
details_list=[]
for pid in random_ids:
  url=BASE_PERSON_URL.format(pid)
  print("Downloading", url)
  person_json=json.loads(requests.get(url).text)
  details_list.append({'id':pid,"details":person_json})
  time.sleep(random.uniform(1,3))
```

##Extract facts into DataFrame

```python
records=[]
for item in details_list:
  pid=item["id"]
  details=item["details"]
  facts=details.get("facts",[])
  if isinstance(facts, dict):
    facts=list(facts.values()) #convert into list

  for f in facts:
    records.append({
        "person_id":pid,
        "fact_type":f.get("type"),
        "date":f.get("date"),
        "place":f.get("place"),
        "comment":f.get("comment")
    })

df_facts=pd.DataFrame(records)
df_facts.head()
```

##4. Write into one CSV file

```python
combined_df=df_people.merge(df_facts,left_on="id",right_on="person_id",how="left")
combined_df=combined_df.merge(df_names,left_on="id",right_on="id",how="left")
combined_df.to_csv(pathlib.Path(EX_DST_PATH, "combined_df.csv"))
```

##5. Write the to an SQLite3 database.

```python
import sqlite3
```

```python
conn=sqlite3.connect(pathlib.Path(EX_DST_PATH, "combined_df.db"))
cur=conn.cursor()
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS people(
  id INTEGER PRIMARY KEY,
  names TEXT

);
""")
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS names(
  person_id INTEGER,
  full_name TEXT,
  given TEXT,
  surname TEXT
)
""")
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS facts (
  person_id INTEGER,
  fact_type TEXT,
  date TEXT,
  place TEXT,
  comment TEXT
)
""")
```

##Insert data

```python
df_people_to_sql = df_people[["id", "names"]].copy()
df_people_to_sql["names"] = df_people_to_sql["names"].apply(json.dumps)
df_people_to_sql.to_sql("people", conn, if_exists="append", index=False)
```

```python
df_names.rename(columns={"id":"person_id","names":"full_name"})[["person_id","full_name","given","surname"]].to_sql("names",conn,if_exists="append",index=False)
```

```python
df_facts.to_sql("facts",conn,if_exists="append",index=False)
conn.commit()
conn.close()
print("SQLite database saved")
```
